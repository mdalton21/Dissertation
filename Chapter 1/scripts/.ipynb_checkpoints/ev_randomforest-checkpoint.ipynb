```python
# Installing packages
import numpy as np
import pandas as pd
import seaborn as sns
import sklearn
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings('ignore')

## Packages for RF and performance analyses
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split, cross_val_score, learning_curve, validation_curve
from sklearn.inspection import permutation_importance
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import roc_curve, roc_auc_score, precision_recall_curve, auc, confusion_matrix, classification_report, accuracy_score
from sklearn.ensemble import RandomForestClassifier  # For classification
from sklearn.impute import SimpleImputer
from sklearn.pipeline import Pipeline
from pathlib import Path

## directory/file paths
base = Path.home() / "/Users/mdalton/Library/CloudStorage/OneDrive-ThePennsylvaniaStateUniversity"
path = base / "Dissertation" / "(1) EV Strategies" / "data" / "df_civilians.csv"
```

```python
# Load in and clean dataset
data = pd.read_csv(path)

# Drop unneeded columns
civ_data = data.drop(columns=['ActorNameNew','ActorTypeNew','country','year'])
#print(civ_data)
```

```python
# Define Xs and Y
X = civ_data.drop('elect_vio', axis=1)  #Means that the x variable is everything in the EMP dataset except attrition
y = civ_data['elect_vio'] # The dependent variable is attrition
```

## Random Forest

```python
# Step 1: Scaling predictors & test/train split
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
X_scaled_df = pd.DataFrame(X_scaled, columns= X.columns)

X_train, X_test, y_train, y_test = train_test_split(
    X_scaled_df, y, test_size=0.2, random_state = 42) 

# Step 2: Fit Random Forest 
clf = RandomForestClassifier(n_estimators=100, random_state=42) # 100 decision trees
clf.fit(X_train, y_train) # Training the algorithm using the training set

# Step 3: Predicted probabilities 
y_pred = clf.predict(X_test) 
y_pred_prob = clf.predict_proba(X_test)[:, 1]
```

### Assessing Performance

```python
#Step 4: Calculating CV accuracy scores
scores = cross_val_score(estimator=clf,
                         X=X_train,
                         y=y_train,
                         cv=10,
                         n_jobs=1)

print(f'Accuracy scores: {scores}')
print(f'Accuracy: {np.mean(scores):.3f} '
      f'+/- {np.std(scores):.3f}')

report = classification_report(y_test, y_pred)
print(report)
```

```python
# Step 5: ROC and AUC
## ROC AUC
fpr_rf, tpr_rf, thresholds = roc_curve(y_test, y_pred_prob)
rf_roc_score = roc_auc_score(y_test, y_pred_prob)

## Plot
plt.title(' ')
plt.plot(fpr_rf, tpr_rf, 'b', label = 'ROC AUC = %0.3f' % rf_roc_score)
plt.legend(loc = 'lower right')
plt.plot([0, 1], [0, 1], 'k--')
plt.xlim([-0.05, 1.05])
plt.ylim([-0.05, 1.05])
plt.ylabel('True Positive Rate')
plt.xlabel('False Positive Rate')
#plt.savefig('rf_roc.png')
plt.show()

print(f'Area under the ROC curve: {rf_roc_score:.5f}')
```

```python
# Step 6: Calculate permutation importance & plot
perm_importance = permutation_importance(clf, X_test, y_test)
feature_perm = perm_importance.importances_mean # Access the importance scores

print(feature_perm)

# Step 7: Graphing most important features
sorted_importances_idx = perm_importance.importances_mean.argsort()
importances = pd.DataFrame(
    perm_importance.importances[sorted_importances_idx].T, 
    columns=X.columns[sorted_importances_idx],)

mean_importances_rf = importances.mean()  # Series: index=feature names, values=mean importance
mean_importances_rf = mean_importances_rf.sort_values()  # sort for nicer plotting

fig, ax = plt.subplots(figsize=(10, 6))
ax.scatter(mean_importances_rf.values, mean_importances_rf.index, color='black')
ax.axvline(x=0, color='k', linestyle='--')
ax.set_xlabel("Decrease in accuracy score")
ax.set_ylabel("")
ax.set_title("")
plt.tight_layout()
#plt.savefig('rf_perm.png')
plt.show()
```

## Neural Network

```python
## packages for NN modeling
import random
import tensorflow as tf
from sklearn.model_selection import StratifiedKFold, cross_val_score
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from tensorflow.keras.optimizers import Adam
from tensorflow.keras.callbacks import History
from scikeras.wrappers import KerasClassifier
```

```python
random.seed(42)
np.random.seed(42)
tf.random.set_seed(42)

## Step 1: Function to build NN model
def build_nn(input_dim=None, **kwargs):
    # Step 1a: Define NN
    nn = Sequential([
        Dense(32, activation='relu', input_shape=(input_dim,)),
        Dropout(0.2),                # optional regularization
        Dense(16, activation='relu'),
        Dropout(0.2),
        Dense(1, activation='sigmoid')
    ])
    # Step 1b: Compile Model
    nn.compile(
        optimizer=Adam(learning_rate=0.001),
        loss='binary_crossentropy',
        metrics=['accuracy']
    )
    # Step 1c: Return Model
    return nn

# Step 2: Split data
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.3, random_state=42
)

# Step 3: Create Keras wrapper
history_callback = History()  # callback to capture metrics

nn_wrapped = KerasClassifier(
    build_fn=build_nn,
    input_dim=X_train.shape[1],
    epochs=50,
    batch_size=32,
    verbose=1,                  
    callbacks=[history_callback]  # record history
)

# Step 4: Build the pipeline
pipeline = Pipeline([
    ('imputer', SimpleImputer(strategy='median')),
    ('scaler', StandardScaler()),
    ('nn', nn_wrapped)
])


# Step 5: Fit pipeline
pipeline.fit(X_train, y_train)
```

```python
# Step 7: Predicted probabilities 
y_pred = pipeline.predict(X_test)
y_pred_prob = pipeline.predict_proba(X_test)[:, 1]
print(classification_report(y_test, y_pred))

# Step 8: CV accuracy
scores = cross_val_score(
    estimator=pipeline,
    X=X_train,
    y=y_train,
    cv=10,
    n_jobs=1
)

print(f"Accuracy scores: {scores}")
print(f"Accuracy: {np.mean(scores):.3f} +/- {np.std(scores):.3f}")
```

```python
# Step 8: ROC and AUC
## Calculate ROC AUC
fpr_nn, tpr_nn, thresholds_nn = roc_curve(y_test, y_pred_prob)
nn_roc_score = roc_auc_score(y_test, y_pred_prob)

print(f'Area under the ROC curve: {nn_roc_score:.5f}')

# Step 9: Calculate permutation importance & plot
perm_importance = permutation_importance(pipeline, X_test, y_test)
feature_perm = perm_importance.importances_mean # Access the importance scores

sorted_importances_idx = perm_importance.importances_mean.argsort()
importances = pd.DataFrame(
    perm_importance.importances[sorted_importances_idx].T, 
    columns=X.columns[sorted_importances_idx],)

mean_importances_nn = importances.mean()  # Series: index=feature names, values=mean importance
mean_importances_nn = mean_importances_nn.sort_values()  # sort for nicer plotting
```

## Logit Model

```python
# Step 1: Create an imputer to replace NaNs 
imputer = SimpleImputer(strategy='mean')

# Step 2: Create logistic regression & pipeline
logreg = LogisticRegression(max_iter=1000)
pipeline = Pipeline([
    ('imputer', imputer),
    ('logreg', logreg)
])

# Step 3: Train-test split
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.3, random_state=42, stratify=y
)

# Step 4: Fit the pipeline
pipeline.fit(X_train, y_train)

# Step 5: Predict probabilities
y_pred_prob = pipeline.predict_proba(X_test)[:, 1]
y_pred = pipeline.predict(X_test)
```

```python
#Step 6: Calculating accuracy scores
scores = cross_val_score(estimator=pipeline,
                         X=X_train,
                         y=y_train,
                         cv=10,
                         n_jobs=1)

print(f'Accuracy scores: {scores}')
print(f'Accuracy: {np.mean(scores):.3f} '
      f'+/- {np.std(scores):.3f}')

report = classification_report(y_test, y_pred)
print(report)
```

```python
# Step 7: ROC and AUC
fpr_lr, tpr_lr, thresholds_lr = roc_curve(y_test, y_pred_prob)
lr_roc_score = roc_auc_score(y_test, y_pred_prob)

print(f'Area under the ROC curve: {lr_roc_score:.5f}')

# Step 8: Feature importance 
coefficients = logreg.coef_[0]
feature_names = X.columns

coef_df = pd.DataFrame({
    'feature': feature_names,
    'coef': coefficients
}).sort_values(by='coef', key=abs, ascending=False)  # sort by absolute value

# Plot
fig, ax = plt.subplots(figsize=(10, 6))
ax.scatter(coef_df['coef'], coef_df['feature'], color='black')  # x=coefficients, y=feature names
ax.axvline(x=0, color='k', linestyle='--')
ax.set_xlabel("Coefficient value / Decrease in accuracy")
ax.set_ylabel("")  # features are already on y-axis
#ax.set_title("Logistic Regression Feature Importance")
plt.tight_layout()
plt.show()
```
