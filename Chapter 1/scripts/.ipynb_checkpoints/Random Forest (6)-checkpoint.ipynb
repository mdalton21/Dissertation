 <span style="font-family: Calibri">The EmployeeAtt.csv dataset contains data on employee attrition rates, a binary variable (yes/no), and other work related variables such as age, travel frequency, income, etc. In this tutorial, we will compare SVM and random forest algorithm to predict generally why employees leave organizations.</span>)

```python
# Installing packages
import numpy as np
import pandas as pd
import seaborn as sns
import mlxtend
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.svm import SVC
from sklearn.inspection import permutation_importance
from mlxtend.plotting import plot_decision_regions
from sklearn.ensemble import RandomForestClassifier  # For classification
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score  # For classification accuracy score
from sklearn.metrics import classification_report 
```

```python
# Naming the dataset
data = pd.read_csv('EmployeeAtt.csv')
```

```python
#Transforming variables 
data["Attrition"] = data["Attrition"].replace({"No": -1, "Yes": 1})
print(data)
```

```python
#Dropping variables

EM= data.drop(columns=['Department','EducationField','JobRole','Over18', 'Gender', 'MaritalStatus','BusinessTravel'])

EMP = EM.dropna(how='all') #Dropping rows with missing variables 
print(EMP)
```

# <span style="font-family: Pacifico, cursive; color: Green"> Data Distributions</span>

Before we start training the models, let's see how the data looks

```python
#Frequency distribution

A_freq_dist = data['Attrition'].value_counts()

print(A_freq_dist)
```

```python
#Histogram of monthly Income

plt.hist(data['MonthlyIncome'], bins=5, edgecolor='k')  # Bins is the number of groups
plt.xlabel('Monthly Income')
plt.ylabel('Frequency')
plt.title('Histogram of Monthly Income')
plt.show()

#Histogram of Age
plt.hist(data['Age'], bins=5, edgecolor='k')  
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.title('Histogram of Age')
plt.show()

#Histogram of job satisfaction
plt.hist(data['JobSatisfaction'], bins=4, edgecolor='k') 
plt.xlabel('Job Satisfaction')
plt.ylabel('Frequency')
plt.title('Histogram of Job Satisfaction')
plt.show()
```

```python
# Calculate the correlation matrix
correlation_matrix = EMP.corr()

# Create a heatmap
plt.figure(figsize=(12, 10))
sns.heatmap(correlation_matrix, annot=True, cmap='RdBu_r', fmt=".2f", linewidths=0.5)
plt.title('Correlation Matrix')
plt.show()
```

## <span style="font-family: Comic Sans MS; color: Green"> Support Vector Machine</span>

<span style="font-family: Calibri">Now let's try predicting employee attrition using Support Vector Machine</span>

```python
EMP.dropna(subset=['MonthlyIncome', 'Age', 'JobSatisfaction', 'Attrition'], inplace=True)  # Drop rows with missing values
print(EMP)

X = EMP.drop('Attrition', axis=1)  #Means that the x variable is every variable except Attrition
y = EMP['Attrition'] # The target variable is Attrition
```

```python
# Standardizing predictors

scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
X_scaled_df = pd.DataFrame(X_scaled, columns= X.columns)
```

```python
# Testing whether the standardization worked 

# Calculate the mean and standard deviation of each feature
mean_values = np.mean(X_scaled_df, axis=0)
std_dev_values = np.std(X_scaled_df, axis=0)

print("Mean values of standardized features:")
print(mean_values)

print("\nStandard deviation values of standardized features:")
print(std_dev_values)
```

```python
#Identifying if the samples are the same size, SVM cannot run if the samples are different sizes

print("X_scaled_df shape:", X_scaled_df.shape)
print("y shape:", y.shape)
```

```python
#SVM code 

X_train, X_test, y_train, y_test = train_test_split(X_scaled_df, y, test_size=0.3, random_state=42)

# Create an SVM model (in this case, a linear SVM)
svm = SVC(kernel = 'linear', C=1.0, random_state=1)

# Train the SVM model on the training data
svm.fit(X_train, y_train)

# Make predictions on the test data
y_pred = svm.predict(X_test)

# Evaluate the model
accuracy = accuracy_score(y_test, y_pred)
print("Accuracy:", accuracy)

# Print a classification report
report = classification_report(y_test, y_pred)
print("\nClassification Report:\n", report)
```

<span style="font-family: Calibri; color: Green">**Overall the model has high accuracy, however we can see that the model does not predict employees who are likely to leave with high precision or recall. Let's find the most important features of the model**</span>


```python
# Calculate permutation importance
perm_importance = permutation_importance(svm, X_test, y_test)

# Access the importance scores
feature_perm = perm_importance.importances_mean

print(feature_perm)
```

```python
#Graphing the most important features

sorted_importances_idx = perm_importance.importances_mean.argsort()
importances = pd.DataFrame(
    perm_importance.importances[sorted_importances_idx].T,
    columns=X.columns[sorted_importances_idx],)

SVMP = importances.plot.box(vert=False, whis=10)
SVMP.set_title("Permutation Importances (test set)")
SVMP.axvline(x=0, color="k", linestyle="--")
SVMP.set_xlabel("Decrease in accuracy score")
SVMP.figure.tight_layout()
```

<span style="font-family: Calibri; color: Green">**Now I will create a graph using the top two features**</span>

```python
#Reloading data in for ease

data = pd.read_csv('EmployeeAtt.csv')

data["Attrition"] = data["Attrition"].replace({"No": 0, "Yes": 1}) #Data has to be binary coded to create graph
print(data)
```

```python
#Redoing the SVM with only 2 variables age and monthly income

data.dropna(subset=['OverTime', 'TotalWorkingYears', 'Attrition'], inplace=True)  # Drop rows with missing values

X = data[['OverTime', 'NumCompaniesWorked']]  #Means that the x variable is only these two variables
y = data['Attrition'] # The dependent variable is attrition

scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
X_scaled_df = pd.DataFrame(X_scaled, columns= X.columns)

X_train, X_test, y_train, y_test = train_test_split(X_scaled_df, y, test_size=0.3, random_state=42)

# Create an SVM model (in this case, a linear SVM)
svm = SVC(kernel = 'linear', C=1.0, random_state=1)

# Train the SVM model on the training data
svm.fit(X_train.values, y_train)

# Make predictions on the test data
y_pred = svm.predict(X_test)

accuracy = accuracy_score(y_test, y_pred)
print("Accuracy:", accuracy)

# Print a classification report
report = classification_report(y_test, y_pred)
print("\nClassification Report:\n", report)
```

<span style="font-family: Calibri; color: Green">**This model is incredibly poor at predicting employees with intentions to leave, but this is a great opportunity to learn what a poor predicting SVM looks like in graph form!**</span>

```python
#Graphing the variables
X_plot = np.vstack((X_train.values, X_test.values))

# Combine y_train and y_test for plotting
y_plot = np.hstack((y_train.values, y_test.values))

# Plot the decision regions
plot_decision_regions(X=X_plot, y=y_plot, clf=svm, legend=2)
plt.xlabel('Monthly Income')
plt.ylabel('Age')
plt.legend(loc='upper left')
plt.tight_layout()
plt.show()
```

# <span style="font-family: Comic Sans MS; color: Green"> Random Forrest</span>

<span style ="font-family: Calibri" > Now we will try to predict employee attrition using a random forest algorithm</span> 

```python
# Naming the dataset
data = pd.read_csv('EmployeeAtt.csv')
```

```python
#Defining X & Y

data["Attrition"] = data["Attrition"].replace({"No": -1, "Yes": 1})
print(data)
```

```python
EMP= data.drop(columns=['Department','EducationField','JobRole','Over18', 'Gender', 'MaritalStatus','BusinessTravel'])
print(EMP)
```

```python
#Defining x and y

X = EMP.drop('Attrition', axis=1)  #Means that the x variable is everything in the EMP dataset except attrition
y = EMP['Attrition'] # The dependent variable is attrition

#Scaling predictors

scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
X_scaled_df = pd.DataFrame(X_scaled, columns= X.columns)

#training dataset

X_train, X_test, y_train, y_test = train_test_split(X_scaled_df, y, test_size=0.2, random_state = 42) #splitting the dataset:20% for testing, 80% for training 
```

```python
#random forest code

clf = RandomForestClassifier(n_estimators=100, random_state=42) # n_estimators determines the number of decision trees that will be created
clf.fit(X_train, y_train) #Training the algorithm using the training set
y_pred = clf.predict(X_test) # predicting attrition outcomes from the x variables in the testing set
```

```python
#Calculating accuracy scores

accuracy = accuracy_score(y_test, y_pred)
print(f'Accuracy: {accuracy:.2f}')

report = classification_report(y_test, y_pred)
print(report)
```

<span style=" font-family: Calibri; color:Green" > **Similar to the SVM, the accuracy of the model is relatively high, however the precision and recall rate are low for predicting employees with intentions to leave. Let's see the most important features of the model**</span> 

```python
# Calculate permutation importance
perm_importance = permutation_importance(clf, X_test, y_test)

# Access the importance scores
feature_perm = perm_importance.importances_mean

print(feature_perm)
```

```python
#Graphing the most important features

sorted_importances_idx = perm_importance.importances_mean.argsort()
importances = pd.DataFrame(
    perm_importance.importances[sorted_importances_idx].T,
    columns=X.columns[sorted_importances_idx],)

SVMP = importances.plot.box(vert=False, whis=10)
SVMP.set_title("Permutation Importances (test set)")
SVMP.axvline(x=0, color="k", linestyle="--")
SVMP.set_xlabel("Decrease in accuracy score")
SVMP.figure.tight_layout()
```

<span style=" font-family: Comic Sans MS; color:Green" > Nearest Neighbors </span> 

```python
# Naming the dataset
data = pd.read_csv('EmployeeAtt.csv')
```

```python
#Defining X & Y

data["Attrition"] = data["Attrition"].replace({"No": -1, "Yes": 1})
print(data)
```

```python
#Dropping variables in dataset

EMP= data.drop(columns=['Department','EducationField','JobRole','Over18', 'Gender', 'MaritalStatus','BusinessTravel'])
print(EMP)
```

```python
EMP.dropna(subset=['MonthlyIncome', 'Age', 'JobSatisfaction', 'Attrition'], inplace=True)  # Drop rows with missing values
print(EMP)

X = EMP.drop('Attrition', axis=1)  #Means that the x variable is everything in the EMP dataset except attrition
y = EMP['Attrition'] # The dependent variable is attrition
```

```python
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
X_scaled_df = pd.DataFrame(X_scaled, columns= X.columns)

print("X_scaled_df shape:", X_scaled_df.shape)
print("y shape:", y.shape)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X_scaled_df, y, test_size=0.3, random_state=42)

# Convert data to NumPy arrays
X_train = np.array(X_train)
X_test = np.array(X_test)

knn = KNeighborsClassifier(n_neighbors=5)
knn.fit(X_train, y_train)

y_pred = knn.predict(X_test)
```

```python
accuracy = accuracy_score(y_test, y_pred)
print(f'Accuracy: {accuracy:.2f}')

report = classification_report(y_test, y_pred)
print(report)
```

```python
#Finding the most important features

perm_importance = permutation_importance(knn, X_test, y_test)

# Access the importance scores
feature_perm = perm_importance.importances_mean

print(feature_perm)
```

```python
#Graphing the most important features

sorted_importances_idx = perm_importance.importances_mean.argsort()
importances = pd.DataFrame(
    perm_importance.importances[sorted_importances_idx].T,
    columns=X.columns[sorted_importances_idx],)

SVMP = importances.plot.box(vert=False, whis=10)
SVMP.set_title("Permutation Importances (test set)")
SVMP.axvline(x=0, color="k", linestyle="--")
SVMP.set_xlabel("Decrease in accuracy score")
SVMP.figure.tight_layout()
```

<span style=" font-family: Calibri; color:Green" > **Comparing all the algorithms together, the SVM algorithm has the best accuracy, and also the best precision and recall rate for employees with intentions to leave. The permutation importance code for each of the models share the analysis that Overtime is an important feature for predicting employee intentions to stay.**</span>
